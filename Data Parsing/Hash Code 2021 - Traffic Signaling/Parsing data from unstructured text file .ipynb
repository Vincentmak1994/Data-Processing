### Parse Input Data

```python
input_data = 'hashcode.in'
```

Let's take a look at the file
#### The first line contains five numbers: duration of simulation, # of of intersections, # of streets, # of cars, bonus points for each car
#### The next S (# of streets) lines contain descriptions of streets
#### The next V (# of cars) lines describe the paths of each car.

```python
open(input_data).readlines()
```

```python
def parse_data(input_data):
    with open(input_data) as f:
        arr = f.readlines()
    
    #Extract general info -> duration, # of interactions, # of streets, # of cars, score
    arr = arr[::-1] #reverse list -> new list order car info -> street info -> general info
    items = arr.pop().split() # -> pop last item (info of duration, # of interactions, # of streets, # of cars, score)
    duration, num_interactions, num_streets, num_cars, score = [int(i) for i in items]
    
    #Extract street info -> format (interaction start, interaction end, street name, travel time)
    #street return format - dict : {street_name: start, end, length}
    streets = {}
    for _ in range(num_streets):
        items = arr.pop().split()
        start = int(items[0])
        end = int(items[1])
        street_name = items[2]
        length = int(items[3])
        streets[street_name] = start,end,length
    
    
    #Extract car info -> format (# of streets car wants to travel, first street, second, etc. )
    #output format - list: [first street, second, etc.]
    
    cars = []
    for _ in range(num_cars):
        items = arr.pop().split()
        path = items[1:] #ignoring first element which is length of path 
        cars.append(path)
    
    return duration, num_interactions, num_streets, num_cars, score, streets, cars
```

```python
duration, num_interactions, num_streets, num_cars, score, streets, cars = parse_data(input_data)
```
